```python
#BTVN
# cho ma trận =[[99,99,99],
#              [99,99,99],
#              [99,99,99]],
#hãy tưởng tượng ô 99 trống
#giả sử 1 là x
# Giả sử 0 là o


#lấy đầu vào từ phía X cà phía O luân phiên
#nếu đầu vào từ phía X là (0,0)

#ma trận sẽ trở thành
#[[x,99,99],
#[99,99,99],
#[99,99,99]]

#Nếu đầu vào từ phía O là (0,0)
#thông báo họ nhập sai, vui lòng nhập lại
#nếu không thì điền vào ma trận

#Thử thách: nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi.
```

```python
# Khởi tạo ma trận 3x3 với giá trị 99 
board = [
    [99, 99, 99],
    [99, 99, 99],
    [99, 99, 99]
]

# Hàm hiển thị bàn cờ cho dễ nhìn
def print_board():
    for row in board:
        # Chuyển đổi 99 thành '.', 1 thành 'X', 0 thành 'O'
        display_row = [('X' if cell == 1 else ('O' if cell == 0 else '.')) for cell in row]
        # tạo danh sách mới hiển thị trạng thái của mỗi ô trong hàng
        print(" | ".join(display_row))
    print("-" * 10)

# Hàm kiểm tra xem đã có người thắng chưa
def check_winner():
    # Kiểm tra hàng ngang và cột dọc
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != 99: return True
        if board[0][i] == board[1][i] == board[2][i] != 99: return True
    
    # Kiểm tra 2 đường chéo
    if board[0][0] == board[1][1] == board[2][2] != 99: return True
    if board[0][2] == board[1][1] == board[2][0] != 99: return True
    
    return False

# Biến điều khiển lượt chơi (True là X, False là O)
is_x_turn = True
turns = 0

print("Chào mừng đến với game Tic-Tac-Toe!")
print_board()

while turns < 9:
    player = "X" if is_x_turn else "O"
    val = 1 if is_x_turn else 0
    
    print(f"Lượt của người chơi {player}")
    
    try:
        # Nhập tọa độ hàng và cột
        row = int(input("Nhập hàng (0, 1, 2): "))
        col = int(input("Nhập cột (0, 1, 2): "))
        
        # Kiểm tra tọa độ hợp lệ
        if row not in [0, 1, 2] or col not in [0, 1, 2]:
            print("Tọa độ không hợp lệ! Vui lòng nhập từ 0 đến 2.")
            continue
            
        # Kiểm tra ô đã có người đi chưa
        if board[row][col] != 99:
            print("Ô này đã có người đi rồi! Vui lòng nhập lại.")
            continue
            
        # Điền vào ma trận
        board[row][col] = val
        print_board()
        turns += 1
        
        # Kiểm tra điều kiện thắng
        if check_winner():
            print(f"Chúc mừng! Người chơi {player} đã thắng!")
            break
            
        # Đổi lượt
        is_x_turn = not is_x_turn
        
    except ValueError:
        print("Vui lòng chỉ nhập số nguyên!")

if turns == 9 and not check_winner():
    print("Hòa rồi!")
    
```

```python
import numpy as np

#Tạo mảng numpy kích thước 150x5 với các giá trị ngẫu nhiên
# [Chiều cao, Cân nặng, Tuổi, Lương, Điểm TB]
data = np.random.rand(150, 5)

# X (4 cột đầu) và y (cột cuối)
X = data[:, :4]  # Lấy tất cả các hàng, từ cột 0 đến cột 3
y = data[:, 4]   # Lấy tất cả các hàng, cột chỉ số 4

#Tách X, y thành tập Train (70%) và Test (30%)
# Tính số lượng hàng cho tập train: 150 * 0.7 = 105
train_size = int(0.7 * len(X))

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]

# Tạo 10 tập dữ liệu X_train không chồng chéo
# Vì X_train có 105 hàng, chia cho 10 tập thì mỗi tập có khoảng 10 phần tử.
# Ta dùng np.array để chia mảng thành các phần bằng nhau.
x_train_subsets = np.array_split(X_train, 10)


print(f"Kích thước X_train: {X_train.shape}")
print(f"Kích thước X_test: {X_test.shape}")
print(f"Số lượng tập con của X_train: {len(x_train_subsets)}")
print(f"Kích thước tập con đầu tiên: {x_train_subsets[0].shape}")
```
