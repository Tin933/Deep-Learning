```python
import torch

# Khởi tạo giá trị x
x = torch.tensor(1.0, requires_grad=True)

# Định nghĩa hàm số
y = 5*x**5 + 6*x**3 - 3*x + 1

# Tính đạo hàm
y.backward()

print(f"Độ dốc của hàm số tại x = 1.0 là: {x.grad.item()}")
```

```python
#BT2 
# Tạo một tensor ban đầu có giá trị là 2
# Định nghĩa hàm số y = x^3 + 2x^2 + 5x + 1
#hãy tính dy/dx tại điểm x = 2 sử dụng PyTorch Autograd
#Dùng phương pháp gradient descent với learning rate alpha = 0.1 để cập nhật giá trị x trong 10 vòng lập
```

```python
import torch

# Tạo tensor ban đầu có giá trị là 2
x = torch.tensor(2.0, requires_grad=True)
learning_rate = 0.1

print("--- Quá trình tối ưu hóa ---")

# Vòng lặp Gradient Descent (10 lần)
for i in range(10):
    # Định nghĩa hàm số
    y = x**3 + 2*x**2 + 5*x + 1
    
    # Tính dy/dx tại x hiện tại
    if x.grad is not None:
        x.grad.zero_()
        
    y.backward()
    
    # In kết quả đạo hàm tại x = 2 ở vòng lặp đầu tiên
    if i == 0:
        print(f"Đạo hàm dy/dx tại x=2 ban đầu là: {x.grad.item()}")

    # Cập nhật giá trị x: x = x - alpha * dy/dx
    with torch.no_grad():
        x -= learning_rate * x.grad
        
    print(f"Vòng lặp {i+1}: x = {x.item():.4f}, y = {y.item():.4f}")

print(f"\nGiá trị x cuối cùng sau 10 vòng lặp: {x.item():.4f}")
```

```python
#BTVN 3:
# Tạo một tập dữ liệu giả lập với x là số giờ học (ngẫu nhiên từ 1- 10) và y là điểm được tính theo công thức
# y = 3x + 5 + noise với noise là một giá trị ngẫu nhiên nhỏ
# Khởi tạp tham số w và b ngẫu nhiên với requires_grad=True
# Tính MSE
# Tính gradient 
# Cập nhật tham số w và b bằng gradient descent với learning rate alpha = 0.01
# Lập ại quá trình trên trong 100 vòng lập để xem sự hội tụ của mô hình 
```

```python
import torch

# Tạo dữ liệu giả lập
torch.manual_seed(42) # Để kết quả giống nhau mỗi lần chạy
x = torch.randint(1, 11, (100, 1), dtype=torch.float32)
noise = torch.randn(100, 1) * 0.5
y_true = 3 * x + 5 + noise

# Khởi tạo tham số w và b
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

learning_rate = 0.01

# Vòng lặp huấn luyện 100 
for epoch in range(100):
    # Dự báo 
    y_pred = x * w + b
    
    # Tính MSE
    loss = torch.mean((y_pred - y_true)**2)
    
    # Tính gradient 
    loss.backward()
    
    # Cập nhật tham số bằng Gradient
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad
        
        # Xóa gradient sau mỗi cập nhật
        w.grad.zero_()
        b.grad.zero_()
    
    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, w = {w.item():.4f}, b = {b.item():.4f}")

print(f"\nKết quả sau cùng: w ≈ {w.item():.2f}, b ≈ {b.item():.2f}")
```
